```python
import time
def monster(xx):
    
    from bs4 import BeautifulSoup
    soup=BeautifulSoup(xx,'html.parser')
    
    title=[]
    for i in soup.find_all("div",class_="job-tittle"):
        title.append(i.find_all("a",target="_blank")[0].text)
        
    company=[]
    for i in soup.find_all("span",class_="company-name"):
        company.append(i.text)
        
    yearofexp=[]
    for i in soup.find_all("div",class_="exp col-xxs-12 col-sm-3 text-ellipsis"):
        yearofexp.append(i.text.strip())
        
    salary=[]
    for i in soup.find_all("div",class_="package col-xxs-12 col-sm-4 text-ellipsis"):
        salary.append(i.text.strip())
        
    import pandas as pd
    table=pd.DataFrame({'Title':title,'Company':company,'Year Of Experience':yearofexp,
             'Salary':salary})
    return table


from selenium import webdriver
import os
path_tounzipfileie="C:/Users/Administrator/Desktop/Python/Data Scraping/geckodriver-v0.25.0-win64"
os.chdir(path_tounzipfileie)
browser=webdriver.Firefox(path_tounzipfileie)   
from bs4 import BeautifulSoup   
browser.get(url)
html=browser.page_source
master={}
master[0]=monster(html)
btns=browser.find_elements_by_class_name("btn-next-prev")
btns[0].click()
for i in range(1,4):
    btns=browser.find_elements_by_class_name("btn-next-prev")
    html=browser.page_source
    master[i]=monster(html)
    for btn in btns:
        if btn.text=="Next":
            btn.click()
            time.sleep(2)
import pandas as pd
master_table=pd.concat([master[i] for i in master],axis=0)
master_table
```
